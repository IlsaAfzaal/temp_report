```python
dic ={}
while True:
  print(".............Birthday App.................")
  print("NOTE: Just Choose integer values")
  print("1. Show Birthday")
  print("2. Add to Birthday List")
  print("3. Exit")

  while True:
    
    try:
        choice = input("Enter your choice: ")
        choice = int(choice)
        break
    except ValueError:
          print("No valid integer! Please try again ...")
  if choice == 1:
    if len(dic.keys()) == 0:
      print("nothing to Show")
    else:
      name = input("Enter to look for Birthday : ")
      birthday = dic.get(name, "No data found")
      print(birthday)
      print("\U0001F973","\U0001F382")
  elif choice == 2:
    name = input("Enter your friend's Name : ")
    date = input("Please Enter date of Birth : ")
    dic[name] = date
    print("Birthday Added")
  elif choice == 3:
    break
  else:
    print("Please Enter a valid Option")

```
